```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df: pd.DataFrame = pd.read_feather('BTCUSDT_since_20190908.ftr')
```

```python
dtypes_series = df.dtypes
for i in df.dtypes.index:
    if (dtypes_series[i]) is np.dtype('O'):
        df[i] = pd.to_numeric(df[i])
```

```python
for column in df.columns:
    if np.issubdtype(df.dtypes[column], np.number):
        df.plot(x='Open time', y=column)
```
